Dividing the dataset between independent and dependent variables

```python
X = data.drop(['wage_per_hour'], axis=1)
Y = data['wage_per_hour']
```

Building the model

Sequential is the easiest way to build mdoels in keras , allows to build mdoels layer by layer .
Dense is the standard layer type works for most cases . In Dense layer all the nodes in the previous layer are being connected with the nodes in the current layer .
Adding 10 nodes in each of the input layer 

```python
from keras.models import Sequential
from keras.layers import Dense 
 
model = Sequential()

#Getting the number of cols in training data
cols = X.shape[1]

#Adding layers 
model.add(Dense(10, activation = 'relu', input_shape = (cols,))) #input_shape specifies the no. of rows and cols , and the comma indicates there can be any amt of rows 
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1)) #contains only 1 nodes , will be used for prediction
```

Compiling the Model - 
Compiling the model takes 2 parameters , optimizers and loss 

Optimizers controls the learning rate of the model. The learning rate determines how fast the optimal weights for the models are calculated . A smaller learning rate lead to more accurate weights , but the time it will take to compute the weights will be longer 

For loss , mean squared error is being used . 

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

Training the model
For training , 5 parameters will be used - training data, target data , validation split , no. of epochs and callbacks 


```python
from keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(patience = 3)

#Train model
model.fit(X, Y, validation_split = 0.2, epochs = 30,
          callbacks = [early_stopping_monitor])
```

```python
new_model = Sequential()

new_model.add(Dense(200, activation = 'relu' , input_shape = (cols,)))
new_model.add(Dense(200, activation = 'relu'))
new_model.add(Dense(200, activation = 'relu'))
new_model.add(Dense(1))

new_model.compile(optimizer = 'adam', loss = 'mean_squared_error')

new_model.fit(X,Y, validation_split = 0.2, epochs = 30, callbacks = [early_stopping_monitor])
```
